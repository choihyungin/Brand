```python
prd_reviews = open("apmall_prd_reviews.txt", 'r')
reviews = prd_reviews.readlines()
```

```python
month_dict = {'2020.02': [], '2020.01': [], '2019.12': [], '2019.11': [], '2019.10': [], 
              '2019.09': [], '2019.08': [], '2019.07': [], '2019.06': [], '2019.05': []}
```

```python
for month in month_dict:
    for top in top30:
        top_num = top.split(" | ")[0]
        i = 0
        for review in reviews:
            review_num = review.split(" | ")[0]
            if top_num == review_num and review.split(" | ")[2][:-3] == month:
                i += 1
        month_dict[month].append(i)   
```

```python
for month in month_dict:
    print(month_dict[month])
```
