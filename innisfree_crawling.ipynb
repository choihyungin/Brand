```python
from selenium import webdriver as wd
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import ActionChains
import time
import re
import emoji
```

```python
inni = open("inni_prd_info.txt", 'w')
```

```python
driver = wd.Chrome(executable_path='/Users/hyunginchoi/Desktop/com/python/chromedriver')
driver.get('https://www.innisfree.com/kr/ko/Product.do?tp=1')
driver.implicitly_wait(10)
time.sleep(1)
```

```python
product_cnt = 0
url = []
prd_max_s = driver.find_element_by_css_selector('#lnbSrchFrm > section > div.contWrap > div.sortCon > p > strong')
prd_max_num= prd_max_s.text.replace("개","")

for j in range(1,(int(prd_max_num)//20+2)):
    main = driver.find_elements_by_css_selector('.pdtOver > a')
    for i in main:
        url.append(i.get_attribute('href'))      
    driver.execute_script("javascript:goPage('%d');" % (j+1))
    time.sleep(1)


for s in url:

    product_info = {}
    product_list =[]
    driver.get('%s' %s)
    time.sleep(0.2)
    try: 
        max_s = driver.find_element_by_css_selector('#contents > div.pdtViewTop > div.pdtViewImg > div.review > span.reviewLength > span')
        max_s = re.findall("\d+", max_s.text)
        max_num = max_s[0] #리뷰 전체 개수   
    except:
        max_num = 0




    driver.implicitly_wait(10)
    txts = driver.find_element_by_css_selector('.pdtViewCont')

    pdtName = txts.find_element_by_css_selector('.pdtName')
    pdtTxt = txts.find_element_by_css_selector('.pdtTxt')
    price = txts.find_element_by_css_selector('.num')
    express_price= txts.find_element_by_css_selector('div.pdtViewOrderTxt > dl:nth-child(2) > dd')
    rev_cnt = max_num

    product_info['prd_name_shop'] = pdtName.text
    product_info['prd_url'] = s #주소 링크
    product_info['prd_desc'] = pdtTxt.text
    product_info['prd_price'] = (price.text.replace(",",""))
    product_info['prd_discount'] = None
    express_price = express_price.text.replace(",","")
    if '이상' not in express_price: #무료배송 조건을 억지로 맞춤
        product_info['prd_shipping'] = None
        product_info['prd_cond'] = '무료배송'
    else:
        shipping = re.findall("\d+", express_price)
        product_info['prd_shipping'] = shipping[0]
        product_info['prd_cond'] = shipping[1]
    try :
        pdtHead = txts.find_element_by_css_selector('.pdtHead')
        product_info['prd_benefit'] = emoji.demojize(pdtHead.text)
    except NoSuchElementException:
        pdtHead = ''
        product_info['prd_benefit'] = pdtHead
    product_info['prd_gift'] = None
    try:
        rev_avg = driver.find_element_by_css_selector('#contents > div.pdtViewTop > div.pdtViewImg > div.review > .reviewNum')
        product_info['rev_avg'] = (float(rev_avg.text)*20)
    except:
        product_info['rev_avg'] = 0
    product_info['rev_cnt'] = int(rev_cnt)
    product_info['com_id'] = 'innisfree' # 외래키 (임시, 받아와야댐)
    product_info['shop_id'] = 'innishop' # 외래키 (임시, 받아와야댐)

    product_list.append(product_info)
    product_cnt += 1
    print('product done',product_cnt)
    i = 1
    for innimall_prd_data in product_list:
        prd_name_shop = innimall_prd_data['prd_name_shop']     # 상품명
        prd_url = innimall_prd_data['prd_url']          # 상품가격
        prd_desc = innimall_prd_data['prd_desc']        # 상세페이지 링크
        prd_price = innimall_prd_data['prd_price']
        prd_discount = innimall_prd_data['prd_discount']
        prd_shipping = innimall_prd_data['prd_shipping']
        prd_cond = innimall_prd_data['prd_cond']
        prd_benefit = innimall_prd_data['prd_benefit']
        prd_gift = innimall_prd_data['prd_gift']
        rev_avg = innimall_prd_data['rev_avg']
        rev_cnt = innimall_prd_data['rev_cnt']
        com_id = innimall_prd_data['com_id']
        shop_id = innimall_prd_data['shop_id']
        inni.write(str(i) + " | " + prd_name_shop + " | " + str(prd_price) + " | " + str(rev_cnt) + " | " + str(rev_avg))
        
##################
    
###################
```

```python
inni.close()
driver.close()
```
