```python
from matplotlib import font_manager, rc
font_fname = '/Library/Fonts/AppleGothic.ttf'
font_name = font_manager.FontProperties(fname=font_fname).get_name()
rc('font', family = font_name)

from konlpy.tag import Twitter; twitter = Twitter()
from collections import Counter
import nltk
import matplotlib.pyplot as plt
```

```python
glow_reviews = open("glowpick_prd_reviews.txt", 'r')
reviews = glow_reviews.readlines()[1:]
```

```python
pos_sent = []

for review in reviews:
    rating = review.split(" | ")[1]
    if rating == '100':
        pos_sent.append(review.split(" | ")[3])
```

```python
neg_sent = []

for review in reviews:
    rating = review.split(" | ")[1]
    if rating == '40' or rating == '20':
        neg_sent.append(review.split(" | ")[3])
```

```python
pos_result = []
rm_morphs = ["Punctuation", "Josa", "Modifier", "KoreanParticle", "Suffix", "Determiner"]

for pos in pos_sent:
    tag = twitter.pos(pos, norm = True, stem = True)
    im = [i for i in tag if i[1] not in rm_morphs]
    pos_result += im
```

```python
neg_result = []
rm_morphs = ["Punctuation", "Josa", "Modifier", "KoreanParticle", "Suffix", "Determiner"]

for neg in neg_sent:
    tag = twitter.pos(neg, norm = True, stem = True)
    im = [i for i in tag if i[1] not in rm_morphs]
    neg_result += im
```

```python
print(len(pos_result))
print(len(neg_result))
```

```python
pos_words = []
for word in pos_result:
    pos_words.append(word[0])
```

```python
neg_words = []
for word in neg_result:
    neg_words.append(word[0])
```

```python
stopwords_list = ["하다", "거", "같다", "너무", "때", "것", "자다", "바르다", "피부"]
```

```python
rm_stopwords = [w for w in pos_words if w not in stopwords_list] #remove stopwords


fdist = nltk.FreqDist(rm_stopwords)
fdist_common = fdist.most_common() #sorted by frequency

rm_freq_one = [j for j in fdist_common if j[1] > 1] #remove frequency = 1

plt.figure(figsize=(25, 10))
fdist.plot(100)
plt.close()
```

```python
rm_stopwords = [w for w in neg_words if w not in stopwords_list] #remove stopwords


fdist = nltk.FreqDist(rm_stopwords)
fdist_common = fdist.most_common() #sorted by frequency

rm_freq_one = [j for j in fdist_common if j[1] > 1] #remove frequency = 1

plt.figure(figsize=(25, 10))
fdist.plot(100)
plt.close()
```
