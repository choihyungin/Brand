### 리뷰 긍정(5점) 부정(3점 이하) 별 빈출 단어 뽑기

```python
from konlpy.tag import Twitter; twitter = Twitter()
from collections import Counter
import nltk
```

```python
from matplotlib import font_manager, rc
font_fname = '/Library/Fonts/AppleGothic.ttf'
font_name = font_manager.FontProperties(fname=font_fname).get_name()
rc('font', family = font_name)
```

```python
prd_reviews = open('apmall_prd_reviews.txt', 'r')

reviews = prd_reviews.readlines()
```

```python
pos_sent = []
neg_sent = []

for review in reviews:
    if len(review) == 1:
        pass
    else:
        
        rating = review.split(" | ")[3]
        text_only = review.split(" | ")[5][:-1]
        if int(rating) == 5:
            pos_sent.append(text_only)
        elif int(rating) == 4:
            pass
        else:
            neg_sent.append(text_only)
```

```python
pos_result= []

for pos in pos_sent:
    tag = twitter.pos(pos, norm = True, stem = True)
    im = [i for i in tag if i[1] == "Noun" or i[1] == "Adjective" or i[1] == "Verb"]
    pos_result += im
```

```python
neg_result= []

for neg in neg_sent:
    tag = twitter.pos(neg, norm = True, stem = True)
    im = [i for i in tag if i[1] == "Noun" or i[1] == "Adjective" or i[1] == "Verb"]
    neg_result += im
```

```python
pos_words = []
for word in pos_result:
    pos_words.append(word[0])
```

```python
neg_words = []
for word in neg_result:
    neg_words.append(word[0])
```

```python
stopwords_list = ["하다", "좋다", "사용", "구매", "같다", "있다", "제품", "자다", "것", "쓰다", "이니스프리", "거", "않다", "되다",
                  "수", "이", "또", "요", "전", "제", "저", "번", "일", "손", "용", "후", "해", "안", "더", "때", "간", "솜"]
```

```python
rm_stopwords = [w for w in pos_words if w not in stopwords_list] #remove stopwords


fdist = nltk.FreqDist(rm_stopwords)
fdist_common = fdist.most_common() #sorted by frequency

rm_freq_one = [j for j in fdist_common if j[1] > 1] #remove frequency = 1

import matplotlib.pyplot as plt
plt.figure(figsize=(30, 10))
fdist.plot(100)
```

```python
rm_stopwords = [w for w in neg_words if w not in stopwords_list] #remove stopwords


fdist = nltk.FreqDist(rm_stopwords)
fdist_common = fdist.most_common() #sorted by frequency

rm_freq_one = [j for j in fdist_common if j[1] > 1] #remove frequency = 1

import matplotlib.pyplot as plt
plt.figure(figsize=(30, 10))
fdist.plot(100)
```
